```python
%%R
require(car)
require(coefplot)
require(arm)
require(asbio)
```

```python
%%R -nd swarming.data
# read data from csv file
swarming.data <- read.csv("swarming-pred-confusion-dataset.csv", header=T)
```

```python
%%R
# replicate and generation are factors
swarming.data$replicate <- factor(swarming.data$replicate)
#swarming.data$generation <- factor(swarming.data$generation)

# generation and swarm density count should be centered and scaled since they have a large range
swarming.data$cent.generation <- scale(swarming.data$generation, center=T, scale=T)
swarming.data$cent.swarm_density_count <- scale(swarming.data$swarm_density_count, center=T, scale=T)
print(summary(swarming.data))
```

### Null model

```python
%%R
# create array with the mean of the centered/scales swarm density count
swarming.data$swarm.mean.cent.swarm_density_count = array(mean(swarming.data$cent.swarm_density_count), length(swarming.data$cent.swarm_density_count))

# null model
swarming.null <- lm(cent.swarm_density_count ~ swarming.data$swarm.mean.cent.swarm_density_count, data=swarming.data)
print(summary(swarming.null))
```

The null model, of course, offers no explanatory power.

### Only predictor: treatment

```python
%%R
# only predictor: treatment
swarming.var1 <- lm(cent.swarm_density_count ~ treatment, data=swarming.data)
print(summary(swarming.var1))
```

Treatment alone seems to be a decent predictor, accounting for about 20% of the variation in the response. Both treatments are significant predictors.

### Only predictor: generation

```python
%%R
# only predictor: generation
swarming.var2 <- lm(cent.swarm_density_count ~ cent.generation, data=swarming.data)
print(summary(swarming.var2))
```

Generation is a pretty weak predictor by itself, only accounting for about 6% of the variation in the response. Nonetheless, generation is a significant predictor.

### Both predictors without interaction terms

```python
%%R
# both predictors without interaction terms
swarming.var1x2woi <- lm(cent.swarm_density_count ~ treatment + cent.generation, data=swarming.data)
print(summary(swarming.var1x2woi))
```

Using both predictors, they account for about 27% of the variation in the response. This is roughly the sum of the explanatory power that the predictors offered individually, which may indicate that there is little (if any) collinearity between the predictors.

### Both predictors with interaction terms

```python
%%R
# both predictors with interaction terms
swarming.var1x2wi <- lm(cent.swarm_density_count ~ treatment + cent.generation + treatment*cent.generation, data=swarming.data)
print(summary(swarming.var1x2wi))
```

With interaction terms added, the predictors account for 32% of the variation in the response, which is the best model explored so far. The interaction between treatment and generation make a significant predictor.

### 3a) For the models with both predictor variables, also determine whether there is any collinearity that you need to worry about. Please show how (and explain why) you chose to use those measures of collinearity.

I decided to use the variance inflation factor (VIF) because it is a built-in function to determine how much the variance of each predictor variance is increased because of collinearity. The square root of the VIF gives a measure of how much larger the standard error is compared to what the standard error would be if the predictor were completely uncorrelated with the other predictor variables. As a rule of thumb, a VIF >= 5 indicates a high level of collinearity between predictors.

```python
%%R
print(sqrt(vif(swarming.var1x2woi)))
```

The VIF for treatment and generation are 1 in the regression without the interaction variable, indicating that the predictors are entirely uncorrelated.

```python
%%R
print(sqrt(vif(swarming.var1x2wi)))
```

In the regression with the interaction variable, the VIF is 1 for treatment (completely uncorrelated) and ~1.4 for both the generation and interaction variables. This indicates that there is a slight collinearity between the generation and interaction variables (as would be expected!), but not so high of a collinearity that it need be worried about.
